```python
def already_loggedIn(id):
    '''
    id: id of the element
    Returns whether the element exists, if it returns:
    True: We need to the loggin
    False: We don't need to login
    '''
    try:
        driver.find_element(By.ID, id)
    except NoSuchElementException:
        return False
    return True
```
